```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
data=pd.read_csv(r"C:\Users\ARUN\Desktop\M_L\spam.csv")
data
```

```python
data.shape
```

```python
data.Category.unique()
```

```python
data['spam']=data.Category.apply(lambda x:1 if x=='spam' else 0)
```

```python
data.head()
```

```python
data=data.drop('Category',axis=1)
data
```

```python
data.head()
```

```python
x = data.Message
y = data.spam
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=45)
```

```python
x_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
x_train_counter=cv.fit_transform(x_train.values)
```

```python
cv.get_feature_names_out()
```

```python
pd.DataFrame(x_train_counter.toarray(),columns=cv.get_feature_names_out())
```

```python
x_train_counter.shape
```

```python
model=MultinomialNB()
model.fit(x_train_counter,y_train)
```

```python
model.score(cv.transform(x_test),y_test)
```

```python
# predictions

emails=['can we get together to watch football game tomorrow?','upto 20% discount on parking,exclusive offer just for you']
```

```python
email_counter=cv.transform (emails)
model.predict(email_counter)
```
