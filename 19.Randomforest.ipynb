```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from word2number import w2n
```

```python
data = pd.read_csv(r"C:\Users\ARUN\Downloads\automobile.csv")
data
```

```python
data.shape
```

```python
data.dtypes
```

```python
data["body-style"].unique()
```

```python
data.fillna(0,inplace= True)
```

```python
data.isna().sum()
```

```python
ab=LabelEncoder()
```

```python
data["company"]=ab.fit_transform(data["company"])
data
```

```python
data["body-style"]=ab.fit_transform(data["body-style"])
data
```

```python
data["engine-type"]=ab.fit_transform(data["engine-type"])
data
```

```python
from word2number import w2n
```

```python
#pip install word2number
```

```python
data['num-of-cylinders']=data['num-of-cylinders'].apply(w2n.word_to_num)
data
```

```python
x = data.drop(["price"],axis=1)
y = data["price"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=6)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
model=RandomForestRegressor(n_estimators=23,random_state=90)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
pred=model.predict(x_test)
```
